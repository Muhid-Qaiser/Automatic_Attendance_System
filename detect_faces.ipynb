# Display the Image

```python
import cv2

img = cv2.imread("C:/Users/HP/Downloads/students.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


plt.imshow(img)
```

# Display Each Face of the Image

```python

faces = RetinaFace.extract_faces(img_path = "C:/Users/HP/Downloads/students.jpeg", align = True)
for face in faces:
  plt.imshow(face)
  plt.show()
```

# Code used to Save the faces in a Folder to Later use (Image Enhancing)

```python
import os

# Create a directory to save the faces
output_dir = "C:/Users/HP/Downloads/extracted_faces"
os.makedirs(output_dir, exist_ok=True)

# Save each face as a .png file
for i, face in enumerate(faces):
    face_file_path = os.path.join(output_dir, f"face_{i+1}.png")
    plt.savefig(face_file_path, bbox_inches='tight', pad_inches=0)  # Save as PNG
    plt.close()  # Close the plot to free memory
    
    plt.imshow(face)
    plt.axis('off')  # Turn off axis

print(f"Extracted {len(faces)} faces and saved them to '{output_dir}'")
```
